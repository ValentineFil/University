4. Для целевой и каждого из признаков постройте парную регрессионную модель. Визуализируйте линию регрессии и исходные данные на графике. При помощи коэффициентов оцените силу связи.


```python
def linear(x_train, y_train, x_test, y_test, col='df'):
  regressor = linear_model.LinearRegression()
  regressor.fit(x_train, y_train)
  print('Регрессор:', col)
  print('Коэффициенты регрессии:', regressor.coef_)
  print('Коэффициент детерминации:', regressor.score(x_test, y_test))
  y_pred = regressor.predict(x_test)
  plt.scatter(y_test.index, y_test)
  plt.scatter(y_test.index, y_pred)
  plt.show()
```

```python
for col in x_train.columns:
  x0_train = np.array(x_train[col]).reshape(-1, 1)
  x0_test = np.array(x_test[col]).reshape(-1, 1)
  linear(x0_train, y_train, x0_test, y_test, col=col)
```

5. Постройте модель множественной линейной регрессии целевой переменной от всех показателей. Оцените качество регрессии, сделайте вывод.


```python
reg = linear_model.LinearRegression()
reg.fit(x_train, y_train)
print('Коэффициенты регрессии:', reg.coef_)
print('Коэффициент детерминации:', reg.score(x_train, y_train))
```

```python
y_pred = reg.predict(x_test).sum(axis=1)
```

```python
df_y_y = y_test
df_y_y['y_pred'] = y_pred; df_y_y
```

```python
plt.scatter(x = df_y_y.index, y = df_y_y['y'])
plt.scatter(x = df_y_y.index, y = df_y_y['y_pred'])
plt.show()
```
